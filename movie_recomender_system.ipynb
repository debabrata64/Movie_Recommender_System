```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies_db = movies.merge(credits,on='title')
```

```python
movies_db.head(1)
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies_db = movies_db[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_db.head(1)
```

```python
movies_db.isnull().sum()
```

```python
movies_db.dropna(inplace=True)

movies_db.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies_db['genres'] = movies_db['genres'].apply(convert)
movies_db.head()
```

```python
movies_db['keywords'] = movies_db['keywords'].apply(convert)
movies_db.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies_db['cast'] = movies_db['cast'].apply(convert3)
movies_db.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_db['crew'] = movies_db['crew'].apply(fetch_director)
movies_db.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_db['cast'] = movies_db['cast'].apply(collapse)
movies_db['crew'] = movies_db['crew'].apply(collapse)
movies_db['genres'] = movies_db['genres'].apply(collapse)
movies_db['keywords'] = movies_db['keywords'].apply(collapse)
```

```python
movies_db['overview'] = movies_db['overview'].apply(lambda x:x.split())
movies_db.head()
```

```python
movies_db['tags'] = movies_db['overview'] + movies_db['genres'] + movies_db['keywords'] + movies_db['cast'] + movies_db['crew']
```

```python
new_df = movies_db.drop(columns=['overview','genres','keywords','cast','crew'])
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() 
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```

```python
import pickle

pickle.dump(new_df.to_dict(),open('movie_dict.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
